# Choose by Condition

## Dependencies and Imports

```kotlin
import org.ntakt.*
```

```kotlin
val n = 101
val x = DoubleArray(n) { it.toDouble() - (n / 2.0) }
val X = ntakt.doubles(n) { x[it] }
val y = (-X*X / n).exp()
val t = y ge 0.5
val c = t.choose(0.0.asType(), 1.0.asType())
val d = t.choose(y, 0.0.asType().constantRandomAccessibleInterval(t))
```

```kotlin
val dataFrame = mapOf<String, Any>(
    "X" to x + x + x,
    "y" to y.flatIterable.map { it.realDouble } + c.flatIterable.map { it.realDouble } + d.flatIterable.map { it.realDouble },
    "label" to Array(n) { "y" } + Array(n) { "1-0" } + Array(n) { "0-y" },
    "alpha" to DoubleArray(3 * n) { 0.1 }
)
val p = lets_plot(dataFrame) { x = "X"; color = "label"; fill = "label" }
p + 
    // for some reason, geom_area adds up the areas instead of plotting on top of each other
//     geom_area { y = "y"; alpha = "alpha"; fill = "label"; group = "label" } +
    geom_line  { y = "y" } +
    ggsize(800, 500)
```
